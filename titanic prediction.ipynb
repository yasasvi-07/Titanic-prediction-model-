```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
titanic['age_cat'] = titanic['age'].apply(category_age)

titanic['age_cat'].all() 
titanic['age_cat'].any() 
```
